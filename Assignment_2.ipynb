```python
# Histogram for each feature


data.drop(columns=['Serial No.']).hist(bins=10, figsize=(15, 10), color='burlywood', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()    
```

```python
# KDE plots for each feature

plt.figure(figsize=(15, 15))
for i, column in enumerate(data.columns, 1):
    plt.subplot(3, 3, i)  # Adjust grid size if needed
    sns.kdeplot(data[column], color='MEDIUMSEAGREEN')    
    plt.title(f'Distribution of {column}')      
```

```python
# Histogram with KDE overlay for each feature

plt.figure(figsize=(15, 15))
for i, column in enumerate(data.columns, 1):
    plt.subplot(3, 3, i)  # Adjust grid size if needed
    sns.histplot(data[column], bins=10, color='indianred', kde=True, edgecolor='black')   
    plt.title(f'Distribution of {column}')
```

```python
# Overall histogram visualization
                                              
data.hist(figsize = (16,9))
```

```python
# Separate data based on Research participation
research_yes = data[data['Research'] == 1]  # Students who did research
research_no = data[data['Research'] == 0]  # Students who didn't do research


# GRE Score distribution based on Research participation

plt.figure(figsize=(10, 6))
sns.histplot(research_yes['GRE Score'], bins=20, color='darkcyan', label='Did Research', kde=True, stat="frequency")
sns.histplot(research_no['GRE Score'], bins=20, color='CHOCOLATE', label='Did Not Do Research', kde=True, stat="frequency")


# Add labels and title

plt.title('GRE Score Distribution Based on Research Participation', fontsize=16)
plt.xlabel('GRE Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(fontsize=12)                  
plt.grid(axis='y', linestyle='--', alpha=0.7)                      
plt.show()
```
